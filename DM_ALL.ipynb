```python
                ##Scaling
from sklearn.preprocessing import StandardScaler

                ## Initialize StandardScaler
scaler = StandardScaler()
columns_to_scale = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])
data

# StandardScaler(),MinMaxScaler(),MaxAbsScaler(),RobustScaler()
                
                ## mean scaling -> (x-mean)/(max-min)
columns_to_scale = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']
data1=data[columns_to_scale].mean()
data2=data[columns_to_scale].max()-data[columns_to_scale].min()
data[columns_to_scale]=(data[columns_to_scale]-data1)/data2
data
```

```python
                ## multiple regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

y = data['Species']
X = data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm']]
print(X.shape)

SEED = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

print(regressor.intercept_)
print(regressor.coef_)

      ## for showing co-relation
feature_names = X.columns
model_coefficients = regressor.coef_
coefficients_df = pd.DataFrame(data = model_coefficients, 
                              index = feature_names, 
                              columns = ['Coefficient value'])
print(coefficients_df)


y_pred = regressor.predict(X_test)
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results)

regressor.score(X_test,y_test)
```

```python
             ## evalution of model (error)

from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'Mean absolute error: {mae:.2f}')
print(f'Mean squared error: {mse:.2f}')
print(f'Root mean squared error: {rmse:.2f}')
```

# classifier

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,:-1].values
y = data['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=47)

# Support Vector Machine (SVM)
SVC_model = SVC()
SVC_model.fit(X_train, y_train)
SVC_prediction = SVC_model.predict(X_test)
SVC_accuracy=accuracy_score(SVC_prediction, y_test)
SVC_cm=confusion_matrix(SVC_prediction, y_test)
print(SVC_accuracy)
print(classification_report(y_test, y_test))
print(SVC_cm)

# Logistic Regression
logreg_clf = LogisticRegression()
logreg_clf.fit(X_train, y_train)
logreg_prediction = logreg_clf.predict(X_test)
lr_accuracy=accuracy_score(logreg_prediction, y_test)
lr_cm=confusion_matrix(logreg_prediction, y_test)
print(lr_accuracy)
print(classification_report(y_test, y_test))
print(lr_cm)

# K-Nearest Neighbors
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
knn_y_pred = knn_model.predict(X_test)
knn_accuracy=accuracy_score(knn_y_pred, y_test)
knn_cm=confusion_matrix(knn_y_pred, y_test)
print(knn_accuracy)
print(classification_report(y_test, y_test))
print(knn_cm)

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_y_pred = dt_model.predict(X_test)
dt_accuracy=accuracy_score(dt_y_pred, y_test)
dt_cm=confusion_matrix(dt_y_pred, y_test)
print(dt_accuracy)
print(classification_report(y_test, y_test))
print(dt_cm)

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_y_pred = rf_model.predict(X_test)
rf_accuracy=accuracy_score(y_test, rf_y_pred)
rf_cm=confusion_matrix(rf_y_pred, y_test)
print(rf_accuracy)
print(classification_report(y_test, rf_y_pred))
print(rf_cm)


# Gaussian Naive Bayes
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
gnb_y_pred = gnb_model.predict(X_test)
gnb_accuracy=accuracy_score(y_test, gnb_y_pred)
gnb_cm=confusion_matrix(gnb_y_pred, y_test)
print(gnb_accuracy)
print(classification_report(y_test, gnb_y_pred))
print(gnb_cm)
```

```python
                        ##bar plot
accuracies = [lr_accuracy,knn_accuracy, SVC_accuracy,dt_accuracy, rf_accuracy,gnb_accuracy]

models = ['Logistic reg', 'KNN', 'SVM','Decision tree', 'Random Forest','Gaussian NB']

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd','#8c564b']

plt.bar(models, accuracies, color=colors)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison of Different Models')
plt.xticks(rotation=90)
plt.ylim([0, 1])

# Display the plot
plt.show()
```

```python
# Create a list of confusion matrices and corresponding model names
cms = [lr_cm, dt_cm, knn_cm,SVC_cm, rf_cm,gnb_cm]
model_names = ['Logistic Regression', 'Decision Tree', 'K-Nearest Neighbors', 'SVM','Random Forest','Gaussian NB']
x_axis_labels = ["Positive(1)","Negative(0)"]
y_axis_labels = ["Positive(1)","Negative(0)"]
# Plot the confusion matrices
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
for ax, cm, model_name in zip(axes.flatten(), cms, model_names):
    sns.set(font_scale=1)
    sns.heatmap(cm,xticklabels = x_axis_labels, yticklabels = y_axis_labels, annot=True, fmt='g', ax=ax,annot_kws = {'size': 16})
    ax.set_ylabel('Predicted',fontsize = 20)
    ax.set_xlabel('Actual',fontsize = 20)
    ax.set_title(model_name,fontsize = 25)

plt.tight_layout()
plt.show()
```
